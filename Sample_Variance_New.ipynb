```python
import numpy as np
import itertools as it
popu=[6,3,11,20] #population m=4
p_mean=np.mean(popu)
p_var=np.var(popu) #division by m
print("population mean = {} and population variance = {}".format(p_mean,p_var))
```

```python
import pandas as pd
n=3 #sample size

subsets=list(it.product(popu,repeat=n)) # all subsets of size n with repetition
sample_mean=[]
sample_var_n=[]
sample_var_n_1=[]
for tup in subsets:
    s_mean=np.mean(tup)
    s_var=np.var(tup) #division by n
    s_var_1=np.var(tup, ddof=1) #division by n-1
    sample_mean.append(s_mean)
    sample_var_n.append(s_var)
    sample_var_n_1.append(s_var_1)
        
est_mean=np.mean(sample_mean) #average of sample means
df_1= pd.DataFrame(list(zip(subsets,sample_mean)), columns = ['Subsets', 'Mean'])
print(df_1,'\n')
print("No. of subsets = ",len(subsets) ,'\n')
print("population mean = ",p_mean)
print("average of sample means = {} ".format(est_mean))
```

```python
est_var_n=np.mean(sample_var_n) #average of sample variance, division by n
est_var_n_1=np.mean(sample_var_n_1) #average of sample variance, division by n-1
df_2= pd.DataFrame(list(zip(subsets,sample_var_n,sample_var_n_1)), columns = ['Subsets','S Variance(n)', 'S Variance(n-1)'])
print(df_2,'\n')
print("No. of subsets = ",len(subsets) ,'\n')
print("population variance = ",p_var, '\n')
print("average of sample variances(division by 'n') = {}".format(est_var_n))
print("average of sample variances(division by 'n-1') = {}".format(est_var_n_1))
```
